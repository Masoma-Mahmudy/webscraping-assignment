```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib.pyplot as plt

# scrape Amazon website
url = 'https://www.amazon.com/'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

# identify unique HTML attributes
items = soup.find_all('input')
input_list = []
for item in items:
    input_dict = {}
    input_dict["type"] = item.get("type")
    input_dict["size"] = item.get("size")
    input_list.append(input_dict)

# store data in dataframe
df = pd.DataFrame(input_list)

# plot data
df['type'].value_counts().plot(kind='bar')
plt.title('Input Type Frequencies on Amazon Homepage')
plt.xlabel('Input Type')
plt.ylabel('Frequency')
plt.show()
```
